```python
page_data=BeautifulSoup(driver.page_source,'html.parser')
all_post=(page_data.find_all(class_="v1Nh3"))
```

```python
all_post_url=[]
for i in range(0,10):
    post_url=all_post[i].find("a")
    all_post_url.append('https://www.instagram.com/'+post_url['href'])
all_post_url
```

```python
for i in all_post_url:
    driver.get(i)
    time.sleep(3)
    likecheck=BeautifulSoup(driver.page_source,'html.parser')
    z=likecheck.find(class_='_8-yf5')
    y=z['aria-label']
    if y=="Like":
        print('photo already unliked')
    else:
        print('photo liked')
        like=driver.find_element_by_class_name('fr66n')
        like.click()
```
