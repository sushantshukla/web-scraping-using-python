```python
 url="https://www.thehindu.com/archive/web/2009/11/07/"
data=requests.get(url)
```

```python
data
```

```python
data_b=BeautifulSoup(data.text,"html.parser")
```

```python
data_b
```

```python
link=data_b.find_all(class_="archive-list")
link
```

```python
#list of all articles present on the page
article_link=[]
for i in link:
    article_link.append(i.a.get("href"))
```

```python
 article_link
```

```python
article_link[2]
```

```python
title=[]
category=[]
paragraph=[]
for news_url in article_link:
    data=requests.get(news_url)
    data_b=BeautifulSoup(data.text,"html.parser")
    headline=data_b.h1
    title.append(headline.string.strip())
    type_of_news=data_b.find(class_="section-name")
    category.append(type_of_news.string.strip())
    #body of the news
    # it is inside the p tag
    body_of_the_news=data_b.find('div',id=lambda x: x and x.startswith('content-body-'))
    if body_of_the_news:
        paragraph.append(body_of_the_news.text.strip())
    else:
        paragraph.append(np.nan)
    
```

```python
len(title)
```

```python
len(category)
```

```python
len(paragraph)
```

```python
#create data frame from that lists
title[0:5]
```

```python
category[0:5]
```

```python
dataset=pd.DataFrame({'Headline':title,'body':paragraph,'Topic':category})
dataset
```

```python
dataset.to_csv('HinduNewsScapper.csv')
```
